```python
import random
class Environment:
    def __init__(self):
        self.students_present = False
        self.light_status = "OFF"

    def get_percept(self):
        self.students_present = random.choice([True, False])
        return {"students": self.students_present, "lights": self.light_status}

    def set_lights(self, status):
        self.light_status = status

class ModelBasedAgent:
    def __init__(self):
        self.model = {"prev_students": None, "prev_lights": "OFF"}

    def act(self, percept):
        students = percept["students"]
        lights = percept["lights"]
        action = "No action"

        if students and lights == "OFF":
            action = "Turn lights ON"
        elif not students and lights == "ON":
            action = "Turn lights OFF"

        self.model["prev_students"] = students
        self.model["prev_lights"] = action.split()[-1] if action != "No action" else lights
        return action

def run_agent(agent, env, steps):
    for i in range(1, steps + 1):
        percept = env.get_percept()
        action = agent.act(percept)
        if "ON" in action:
          env.set_lights("ON")
        elif "OFF" in action:
          env.set_lights("OFF")
        print(f"Step {i}:")
        print(f"Percept: Students = {percept['students']}, Lights = {percept['lights']}")
        print(f"Action: {action}")
        print(f"Internal Model: {agent.model}")

env = Environment()
agent = ModelBasedAgent()
run_agent(agent, env, 8)
```

```python
class Environment:
    def __init__(self, subjects):
        self.unfinished_subjects = subjects

    def study_subject(self, subject):
        if subject in self.unfinished_subjects:
            self.unfinished_subjects.remove(subject)
            return True
        return False

    def get_percept(self):
        return self.unfinished_subjects

class GoalBasedAgent:
    def __init__(self, goal):
        self.goal = goal

    def act(self, percept):
        if not percept:
            return None
        return percept[0]

def run_agent(agent, environment):
    while True:
        percept = environment.get_percept()
        subject = agent.act(percept)

        if subject is None:
            print(f"Goal Achieved: {agent.goal}")
            break

        print(f"Studying {subject}")
        environment.study_subject(subject)

subjects_to_study = ["AI", "Math", "Physics"]
env = Environment(subjects_to_study)
agent = GoalBasedAgent("All subjects completed")

run_agent(agent, env)
```

```python
class Environment:
    def __init__(self):
        self.restaurants = [
            {"name": "A", "distance": 3, "rating": 7},
            {"name": "B", "distance": 5, "rating": 9}
        ]

    def get_percept(self):
        return self.restaurants

class UtilityBasedAgent:
    def calculate_utility(self, restaurant):
        return restaurant['rating'] - restaurant['distance']

    def act(self, percept):
        best_restaurant = None
        max_utility = -float('inf')

        for i in percept:
            utility = self.calculate_utility(i)
            print(f"Resturant {i['name']} Utility = {utility}")

            if utility > max_utility:
                max_utility = utility
                best_restaurant = i['name']

        return best_restaurant

def run_agent(agent, environment):
    percept = environment.get_percept()
    selected = agent.act(percept)
    print(f"Selected Restaurant: {selected}")

env = Environment()
agent = UtilityBasedAgent()
run_agent(agent, env)
```

```python
import random

class LearningBasedAgent:
    def __init__(self, actions):
        self.Q = {}
        self.actions = actions
        self.alpha = 0.1
        self.gamma = 0.9
        self.epsilon = 0.1

    def get_Q_value(self, state, action):
        return self.Q.get((state, action), 0.0)

    def select_action(self, state):
        if random.uniform(0, 1) < self.epsilon:
            return random.choice(self.actions)
        else:
            return max(self.actions, key=lambda a: self.get_Q_value(state, a))

    def learn(self, state, action, reward, next_state):
        old_Q = self.get_Q_value(state, action)
        best_future_Q = max([self.get_Q_value(next_state, a) for a in self.actions])
        self.Q[(state, action)] = old_Q + self.alpha * (reward + self.gamma * best_future_Q - old_Q)

    def act(self, state):
        return self.select_action(state)

class GameEnvironment:
    def __init__(self):
        self.state = "Active"

    def get_percept(self):
        return self.state

    def get_reward(self, action):
        if action == 'Play':
            return 5
        else:
            return 1

def run_game(agent, environment, steps):
    for step in range(1, steps + 1):
        state = environment.get_percept()
        action = agent.act(state)
        reward = environment.get_reward(action)
        next_state = state

        agent.learn(state, action, reward, next_state)

        print(f"Step {step}: Action {action} Reward {reward}")

    print("\nQ-table Updated")
    for key, value in agent.Q.items():
        print(f"State-Action {key}: {value:.2f}")

agent = LearningBasedAgent(['Play', 'Rest'])
env = GameEnvironment()
run_game(agent,env,10)
```

```python
class Environment:
    def __init__(self):
        self.grid = ['Safe', 'Safe', 'Fire',
                     'Safe', 'Fire', 'Safe',
                     'Safe', 'Safe', 'Fire']
        self.room_names = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']

    def get_percept(self, position):
        return self.grid[position]

    def extinguish_fire(self, position):
        self.grid[position] = 'Safe'

    def display_grid(self, agent_position):
        print(f"\nRobot at Room: {self.room_names[agent_position]}")
        display = []
        for i in range(len(self.grid)):
            if i == agent_position:
                display.append("R")
            else:
                display.append("!" if self.grid[i] == "Fire" else "*")

        for i in range(0, 9, 3):
            print(" | ".join(display[i:i + 3]))

class FirefightingRobot:
    def __init__(self):
        self.position = 0

    def act(self, percept, environment):
        if percept == 'Fire':
            environment.extinguish_fire(self.position)
            return f"Action: Extinguished fire in room {environment.room_names[self.position]}"
        else:
            return f"Action: Room {environment.room_names[self.position]} is safe"

    def move(self):
        if self.position < 8:
            self.position += 1
            return True
        return False

def run_mission():
    env = Environment()
    robot = FirefightingRobot()

    print("Mission Started")
    print("Legend: R = Robot, ! = Fire, * = Safe")

    for step in range(9):
        percept = env.get_percept(robot.position)
        env.display_grid(robot.position)
        result = robot.act(percept, env)
        print(result)
        if step < 8:
            robot.move()
    print("\nMission Complete: All Rooms Extinguished")
    env.display_grid(robot.position)

run_mission()
```
