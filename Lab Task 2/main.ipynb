Task 1: Vehicle Rental System

```python
class Vehicle:
    def __init__(self,id,brand,rent):
        self.vehicle_id = id
        self.brand = brand
        self.rent_per_day = rent
    def display_details(self):
        print(f"Vehicle Id : {self.vehicle_id}\nBrand : {self.brand}\nRent Per Day : {self.rent_per_day}")
    def calculate_rent(self,days):
        rent=self.rent_per_day * days
        print(f"Total Rent for {days} days is {rent}")
        
v1=Vehicle(1,"Toyota",4000)
v2=Vehicle(2,"Honda",5000)

v1.display_details()
v2.display_details()

v1.calculate_rent(4)
v2.calculate_rent(7)
```

Task 2: Employee Payroll System

```python
class Employee:
    def __init__(self,name,emp_id):
        self.name=name
        self.emp_id=emp_id
    def calculate_salary(self):
        pass
class FullTimeEmployee(Employee):
    def __init__(self, name, emp_id,sal):
        super().__init__(name, emp_id)
        self.monthly_salary=sal
    def calculate_salary(self):
        return self.monthly_salary
class PartTimeEmployee(Employee):
    def __init__(self, name, emp_id,hr_work,hr_rate):
        super().__init__(name, emp_id)
        self.hours_worked=hr_work
        self.hourly_rate=hr_rate
    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked
    
e1= FullTimeEmployee("Saim",808,150000)
e2=PartTimeEmployee("Rafay",866,30,2500)

print("Full Time Employee Salary : ",e1.calculate_salary())
print("Part Time Employee Salary : ",e2.calculate_salary())
```

Task 3: Student Result System

```python
class Student:
    def __init__(self,name):
        self.name=name
        self.__marks=0
    def set_marks(self,marks):
        self.__marks=marks
    def get_marks(self):
        return self.__marks
    def calculate_grade(self):
        if self.__marks >= 85:
            return "A"
        elif self.__marks >= 70:
            return "B"
        elif self.__marks >= 50:
            return "C"
        else:
            return "F"
s1=Student("Saim")
s2=Student("Rafay")
s1.set_marks(92)
s2.set_marks(70)
print("Student Name: ",s1.name," Marks: ",s1.get_marks(),"Grade: ",s1.calculate_grade())
print("Student Name: ",s2.name," Marks: ",s2.get_marks(),"Grade: ",s2.calculate_grade())
```
