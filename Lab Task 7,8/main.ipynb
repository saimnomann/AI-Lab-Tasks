Task 1: Minimax Algorithm Implementation

```python
game_tree = [
  [[4, 7], [2, 5]],
  [[1, 8], [3, 6]]
]

visited_nodes = []

def run_minimax(node, depth, is_max_player):
  if depth == 0 or isinstance(node, int):
    visited_nodes.append(node)
    return node

  if is_max_player:
    best_val = -float('inf')
    for child in node:
      value = run_minimax(child, depth - 1, False)
      if value > best_val:
        best_val = value
    return best_val
  else:
    best_val = float('inf')
    for child in node:
      value = run_minimax(child, depth - 1, True)
      if value < best_val:
        best_val = value
    return best_val

final_score = run_minimax(game_tree, 3, True)

print(f"Root optimal value: {final_score}")
print(f"Sequence of visited nodes: {visited_nodes}")

def run_minimax_limited(node, depth, is_max_player):
  if depth == 0:
    if isinstance(node, list):
      if isinstance(node[0], list):
        return node[0][0]
      else:
        return node[0]
    return node

  if isinstance(node, int):
    return node

  if is_max_player:
    best_val = -float('inf')
    for child in node:
      value = run_minimax_limited(child, depth - 1, False)
      if value > best_val:
        best_val = value
    return best_val
  else:
    best_val = float('inf')
    for child in node:
      value = run_minimax_limited(child, depth - 1, True)
      if value < best_val:
        best_val = value
    return best_val

limit_score = run_minimax_limited(game_tree, 2, True)

print(f"Depth-limited result (Depth 2): {limit_score}")
```

Task 2: Alpha-Beta Pruning

```python
tree = [
  [[4, 7], [2, 5]],
  [[1, 8], [3, 6]]
]

visited_mm = []
visited_ab = []
ab_values = {}
pruned_branches = []

def minimax(node, depth, maximizing):
    if depth == 0 or isinstance(node, int):
        visited_mm.append(node)
        return node

    if maximizing:
        best_val = -1000
        for child in node:
            best_val = max(best_val, minimax(child, depth - 1, False))
        return best_val
    else:
        best_val = 1000
        for child in node:
            best_val = min(best_val, minimax(child, depth - 1, True))
        return best_val

def alphabeta(node, depth, alpha, beta, maximizing, label):
    if depth == 0 or isinstance(node, int):
        visited_ab.append(node)
        return node

    if maximizing:
        best_val = -1000
        for i, child in enumerate(node):
            val = alphabeta(child, depth - 1, alpha, beta, False, label + str(i))
            best_val = max(best_val, val)
            alpha = max(alpha, best_val)

            if beta <= alpha:
                pruned_branches.append(label + " pruned")
                break

        ab_values[label] = best_val
        return best_val
    else:
        best_val = 1000
        for i, child in enumerate(node):
            val = alphabeta(child, depth - 1, alpha, beta, True, label + str(i))
            best_val = min(best_val, val)
            beta = min(beta, best_val)

            if beta <= alpha:
                pruned_branches.append(label + " pruned")
                break

        ab_values[label] = best_val
        return best_val

mm_result = minimax(tree, 3, True)
ab_result = alphabeta(tree, 3, -1000, 1000, True, "Root")

print(f"Optimal value by MiniMax: {mm_result}")
print(f"Optimal value by Alpha-Beta: {ab_result}\n")

print(f"Nodes visited by MiniMax: {len(visited_mm)}")
print(f"Nodes visited by Alpha-Beta: {len(visited_ab)}\n")

print(f"Alpha-Beta node values: {ab_values}")
print(f"Pruned branches during Alpha-Beta: {pruned_branches}")
```

```python
tree = [
    [[4, 7], [2, 5]],
    [[1, 0], [3, 6]]
]

visited_ab = []
pruned_nodes = []
node_values = {}

def minimax(node, depth, maximizing, path_trace):

    if depth == 0 or isinstance(node, int):
        return node, path_trace

    if maximizing:
        best_val = -1000
        best_trace = []
        for i, child in enumerate(node):
            val, trace = minimax(child, depth - 1, False, path_trace + [i])
            if val > best_val:
                best_val = val
                best_trace = trace
        node_values[str(path_trace)] = best_val
        return best_val, best_trace

    else:
        best_val = 1000
        best_trace = []
        for i, child in enumerate(node):
            val, trace = minimax(child, depth - 1, True, path_trace + [i])
            if val < best_val:
                best_val = val
                best_trace = trace
        node_values[str(path_trace)] = best_val
        return best_val, best_trace


def alphabeta(node, depth, alpha, beta, maximizing):

    if depth == 0 or isinstance(node, int):
        visited_ab.append(node)
        return node

    if maximizing:
        best_val = -1000
        for child in node:
            val = alphabeta(child, depth - 1, alpha, beta, False)
            best_val = max(best_val, val)
            alpha = max(alpha, best_val)
            if beta <= alpha:
                pruned_nodes.append("pruned")
                break
        return best_val

    else:
        best_val = 1000
        for child in node:
            val = alphabeta(child, depth - 1, alpha, beta, True)
            best_val = min(best_val, val)
            beta = min(beta, best_val)
            if beta <= alpha:
                pruned_nodes.append("pruned")
                break
        return best_val



minimax_result, optimal_path = minimax(tree, 3, True, [])

ab_result = alphabeta(tree, 3, -1000, 1000, True)

print(f"Minimax computed value: {minimax_result}")
print(f"Optimal path from Minimax: {optimal_path}")
print(f"Branches pruned in Alpha-Beta: {pruned_nodes}")
print(f"Alpha-Beta final value: {ab_result}")
print(f"Nodes visited during Alpha-Beta: {visited_ab}")
```

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()

var_A = model.NewIntVar(0, 3, "A")
var_B = model.NewIntVar(0, 3, "B")
var_C = model.NewIntVar(0, 3, "C")

model.Add(var_A != var_B)
model.Add(var_B != var_C)
model.Add(var_A + var_B <= 4)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status in (cp_model.FEASIBLE, cp_model.OPTIMAL):
    print("A feasible solution has been found:")
    print(f"  A = {solver.Value(var_A)}")
    print(f"  B = {solver.Value(var_B)}")
    print(f"  C = {solver.Value(var_C)}")
else:
    print("No feasible solution exists for the given constraints.")
```

Task 5 Print All Possible Solutions

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()

A = model.NewIntVar(0, 3, "A")
B = model.NewIntVar(0, 3, "B")
C = model.NewIntVar(0, 3, "C")

model.Add(A != B)
model.Add(B != C)
model.Add(A + B <= 4)

class PrintAllSolutions(cp_model.CpSolverSolutionCallback):
    def __init__(self, vars_list):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.vars_list = vars_list
        self.count = 0

    def on_solution_callback(self):
        self.count += 1
        print(f"Solution {self.count}: ", end="")
        for var in self.vars_list:
            print(f"{var.Name()} = {self.Value(var)}", end="  ")
        print()

solver = cp_model.CpSolver()
solver.parameters.enumerate_all_solutions = True

callback = PrintAllSolutions([A, B, C])
solver.Solve(model, callback)

print(f"\nTotal solutions found: {callback.count}")
```

Task 6 Simple Optimization Problem

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()

var_x = model.NewIntVar(0, 20, "x")
var_y = model.NewIntVar(0, 20, "y")
var_z = model.NewIntVar(0, 20, "z")

model.Add(var_x + 2 * var_y + var_z <= 20)
model.Add(3 * var_x + var_y <= 18)

model.Maximize(4 * var_x + 2 * var_y + var_z)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL:
    print(f"  Optimal value for x: {solver.Value(var_x)}")
    print(f"  Optimal value for y: {solver.Value(var_y)}")
    print(f"  Optimal value for z: {solver.Value(var_z)}")
    print(f"  Maximum objective (4x + 2y + z): {solver.ObjectiveValue()}")
else:
    print("No optimal solution exists for the given constraints.")
```

Task 7 N-Queens Variation

```python
from ortools.sat.python import cp_model

N = 4
model = cp_model.CpModel()

queens = [model.NewIntVar(0, N - 1, f"q_row_{i}") for i in range(N)]

model.AddAllDifferent(queens)

for i in range(N):
    for j in range(i + 1, N):
        model.Add(queens[i] - queens[j] != i - j)
        model.Add(queens[j] - queens[i] != i - j)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status in (cp_model.FEASIBLE, cp_model.OPTIMAL):
    print("4-Queens Solution:\n")
    for r in range(N):
        line = ""
        for c in range(N):
            if solver.Value(queens[r]) == c:
                line += " Q "
            else:
                line += " _ "
        print(line)

    print("\nQueen positions:")
    for r in range(N):
        print(f"  Row {r} -> Column {solver.Value(queens[r])}")
else:
    print("No solution found.")
```
