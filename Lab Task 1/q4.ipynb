```python
print("1.Add two numbers\n2.Subtract two number\n3. Exit")
choice=0
result=0
while True:
  choice = int(input("Enter your choice: "))
  if choice==3:
    break
  else:
    num1=int(input("Enter first number: "))
    num2=int(input("Enter second number:" ))
    if choice==1:
      result=num1+num2
    elif choice==2:
      result=num1-num2
    else:
      print("Invalid Choice\n")
      continue
    print("Result: ",result)
```
