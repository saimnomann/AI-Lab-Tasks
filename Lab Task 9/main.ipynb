Task 1

```python
import random

suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
deck = [(rank, suit) for suit in suits for rank in ranks]
print(deck)

red_cards = [card for card in deck if card[1] in ['Hearts', 'Diamonds']]
face_cards = [card for card in deck if card[0] in ['Jack', 'Queen', 'King']]

trials = 100000

red_count = 0
heart_given_red = 0
diamond_given_face = 0
spade_or_queen_given_face = 0

for _ in range(trials):
    card = random.choice(deck)
    if card in red_cards:
        red_count += 1
        if card[1] == 'Hearts':
            heart_given_red += 1

for _ in range(trials):
    card = random.choice(face_cards)
    if card[1] == 'Diamonds':
        diamond_given_face += 1
    if card[1] == 'Spades' or card[0] == 'Queen':
        spade_or_queen_given_face += 1

p_red = red_count / trials
p_heart_given_red = heart_given_red / red_count
p_diamond_given_face = diamond_given_face / trials
p_spade_or_queen_given_face = spade_or_queen_given_face / trials

theoretical_p_red = 26 / 52
theoretical_p_heart_given_red = 13 / 26
theoretical_p_diamond_given_face = 3 / 12
theoretical_p_spade_or_queen_given_face = (3 + 4 - 1) / 12

print(f"Theoretical P(Red): {theoretical_p_red:.3f}")
print(f"Simulated P(Red):   {p_red:.3f}")

print(f"Theoretical P(Heart | Red): {theoretical_p_heart_given_red:.3f}")
print(f"Simulated P(Heart | Red):   {p_heart_given_red:.3f}")

print(f"Theoretical P(Diamond | Face): {theoretical_p_diamond_given_face:.3f}")
print(f"Simulated P(Diamond | Face):   {p_diamond_given_face:.3f}")

print(f"Theoretical P(Spade or Queen | Face): {theoretical_p_spade_or_queen_given_face:.3f}")
print(f"Simulated P(Spade or Queen | Face):   {p_spade_or_queen_given_face:.3f}")
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

model = DiscreteBayesianNetwork([
    ('Intelligence', 'Grade'),
    ('StudyHours', 'Grade'),
    ('Difficulty', 'Grade'),
    ('Grade', 'Pass')
])

cpd_intelligence = TabularCPD(
    variable='Intelligence',
    variable_card=2,
    values=[[0.7], [0.3]],
    state_names={'Intelligence': ['High', 'Low']}
)

cpd_study = TabularCPD(
    variable='StudyHours',
    variable_card=2,
    values=[[0.6], [0.4]],
    state_names={'StudyHours': ['Sufficient', 'Insufficient']}
)

cpd_difficulty = TabularCPD(
    variable='Difficulty',
    variable_card=2,
    values=[[0.4], [0.6]],
    state_names={'Difficulty': ['Hard', 'Easy']}
)

cpd_grade = TabularCPD(
    variable='Grade',
    variable_card=3,
    values=[
        [0.6, 0.4, 0.5, 0.3, 0.8, 0.6, 0.7, 0.5],
        [0.3, 0.4, 0.3, 0.4, 0.15, 0.3, 0.2, 0.3],
        [0.1, 0.2, 0.2, 0.3, 0.05, 0.1, 0.1, 0.2]
    ],
    evidence=['Intelligence', 'StudyHours', 'Difficulty'],
    evidence_card=[2, 2, 2],
    state_names={
        'Grade': ['A', 'B', 'C'],
        'Intelligence': ['High', 'Low'],
        'StudyHours': ['Sufficient', 'Insufficient'],
        'Difficulty': ['Hard', 'Easy']
    }
)

cpd_pass = TabularCPD(
    variable='Pass',
    variable_card=2,
    values=[
        [0.95, 0.80, 0.50],
        [0.05, 0.20, 0.50]
    ],
    evidence=['Grade'],
    evidence_card=[3],
    state_names={
        'Pass': ['Yes', 'No'],
        'Grade': ['A', 'B', 'C']
    }
)

model.add_cpds(cpd_intelligence, cpd_study, cpd_difficulty, cpd_grade, cpd_pass)

assert model.check_model()

infer = VariableElimination(model)

result1 = infer.query(
    variables=['Pass'],
    evidence={'StudyHours': 'Sufficient', 'Difficulty': 'Hard'}
)

print("P(Pass | StudyHours=Sufficient, Difficulty=Hard):")
print(result1)

result2 = infer.query(
    variables=['Intelligence'],
    evidence={'Pass': 'Yes'}
)

print("\nP(Intelligence | Pass=Yes):")
print(result2)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

model = DiscreteBayesianNetwork([
    ('Disease', 'Fever'),
    ('Disease', 'Cough'),
    ('Disease', 'Fatigue'),
    ('Disease', 'Chills')
])

cpd_disease = TabularCPD(
    variable='Disease',
    variable_card=2,
    values=[[0.3], [0.7]],
    state_names={'Disease': ['Flu', 'Cold']}
)

cpd_fever = TabularCPD(
    variable='Fever',
    variable_card=2,
    values=[
        [0.9, 0.5],
        [0.1, 0.5]
    ],
    evidence=['Disease'],
    evidence_card=[2],
    state_names={'Fever': ['Yes', 'No'], 'Disease': ['Flu', 'Cold']}
)

cpd_cough = TabularCPD(
    variable='Cough',
    variable_card=2,
    values=[
        [0.8, 0.6],
        [0.2, 0.4]
    ],
    evidence=['Disease'],
    evidence_card=[2],
    state_names={'Cough': ['Yes', 'No'], 'Disease': ['Flu', 'Cold']}
)

cpd_fatigue = TabularCPD(
    variable='Fatigue',
    variable_card=2,
    values=[
        [0.7, 0.3],
        [0.3, 0.7]
    ],
    evidence=['Disease'],
    evidence_card=[2],
    state_names={'Fatigue': ['Yes', 'No'], 'Disease': ['Flu', 'Cold']}
)

cpd_chills = TabularCPD(
    variable='Chills',
    variable_card=2,
    values=[
        [0.6, 0.4],
        [0.4, 0.6]
    ],
    evidence=['Disease'],
    evidence_card=[2],
    state_names={'Chills': ['Yes', 'No'], 'Disease': ['Flu', 'Cold']}
)

model.add_cpds(cpd_disease, cpd_fever, cpd_cough, cpd_fatigue, cpd_chills)

assert model.check_model()

infer = VariableElimination(model)

result1 = infer.query(
    variables=['Disease'],
    evidence={'Fever': 'Yes', 'Cough': 'Yes'}
)

print("P(Disease | Fever=Yes, Cough=Yes):")
print(result1)

result2 = infer.query(
    variables=['Disease'],
    evidence={'Fever': 'Yes', 'Cough': 'Yes', 'Chills': 'Yes'}
)

print("\nP(Disease | Fever=Yes, Cough=Yes, Chills=Yes):")
print(result2)

result3 = infer.query(
    variables=['Fatigue'],
    evidence={'Disease': 'Flu'}
)

print("\nP(Fatigue | Disease=Flu):")
print(result3)
```

```python
import random

states = ['Sunny', 'Cloudy', 'Rainy']

transition_matrix = {
    'Sunny':   [0.6, 0.3, 0.1],
    'Cloudy':  [0.3, 0.4, 0.3],
    'Rainy':   [0.2, 0.3, 0.5]
}

def next_state(current):
    probs = transition_matrix[current]
    return random.choices(states, probs)[0]

days = 10
trials = 100000
at_least_3_rainy = 0

for _ in range(trials):
    current = 'Sunny'
    rainy_count = 0
    for _ in range(days):
        if current == 'Rainy':
            rainy_count += 1
        current = next_state(current)
    if rainy_count >= 3:
        at_least_3_rainy += 1

probability = at_least_3_rainy / trials

print("Probability of at least 3 rainy days in 10 days:", round(probability, 3))
```
